```python
import pandas as pd

# Creating a DataFrame

data = {'Nome' : ['Marlon', 'Lucas', 'Ana'], "Idade" : [22, 25, 30], 'Salário' :  [1000, 2000, 3000]}
df = pd.DataFrame(data)

df

```

```python
# Seleção por índice rotulado
print(df.loc[0])
```

```python
# Seleção da linha que está na posição 1

print(df.iloc[0])
```

```python
# Filtrando por um critério

filtro = df[df['Idade'] > 23]
filtro
```

```python
# Changing some element of a column

df.loc[0, "Salário"] = 1250
df
```

```python
df["Estado"] = ['SP', 'RJ', 'MG']
df
```

```python
# TO remove a column

df = df.drop('Estado', axis=1)
df
```

```python
# Ordeanando por Idade
df_ordenado = df.sort_values(by="Idade")
df_ordenado
```

```python
# Ordenando descrecentemente

df_ordenado = df.sort_values(by='Idade', ascending = False)
df_ordenado
```

```python
# Para fazer uma operação em toda a coluna

df["Salário"] = df['Salário'].apply(lambda x: x * 2)
df
```

```python
# Checando se há valor nulu

df_nulos = pd.DataFrame({'Nome': ['Alice', 'Julia', None], "Idade" : [25, None, 30]})

print(df_nulos.isna())
```

```python
# Preenchendo os nulos

df_nulos.Idade = df_nulos['Idade'].fillna(df_nulos.Idade.mean())
df_nulos
```

```python
# Removendo os nulos

df_nulos= df_nulos.dropna()
df_nulos
```
